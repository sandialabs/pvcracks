```python
import numpy as np
import json

# this is generated by training/train_channeled_unet_cv_hpo.py
data = json.load(open("nested_cv_results.json"))

accuracies = [
    fold["metrics"]["Aggregate"]["accuracy"]
    for fold in data
]

mean_acc = np.mean(accuracies)
std_acc = np.std(accuracies, ddof=1)  # sample std (correct for CV)

print(f"Mean accuracy: {mean_acc:.6f}")
print(f"Std dev:       {std_acc:.6f}")
```

```python
from scipy import stats

k = len(accuracies)
mean = np.mean(accuracies)
std = np.std(accuracies, ddof=1)

t_val = stats.t.ppf(0.975, df=k-1)
ci_low = mean - t_val * std / np.sqrt(k)
ci_high = mean + t_val * std / np.sqrt(k)

print(f"95% CI: [{ci_low:.6f}, {ci_high:.6f}]")

print(mean)
print(t_val * std / np.sqrt(k))
```
